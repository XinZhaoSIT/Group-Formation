```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('skill_data_label.csv')
df
```

```python
table = pd.pivot_table(df, values=['Teamwork','Empathy_skills','Self-control','Ability_to_compromise',\
                    'Ability_to_teach','Respect_for_others','Openness_to_Criticism','Sensitivity_and_Tact',\
                             'Gaining_trust','Comfort_with_Differences','Listening_Skills'],\
                    index=['label'],
                    aggfunc=[np.mean])
table
```

```python
table.T.plot(figsize=(10,6), \
             xlabel='Interpersonal skills',\
             ylabel='Average score')
plt.xticks(np.arange(11), ['Compromise','Teach',\
                           'Comfort_Diff','Empathy',\
                           'Gaining_trust','Listening',\
                           'Openness_to_Criticism','Respect',\
                           'Self-control','Tact','Teamwork'], rotation=45)
plt.title('Interpersonal Skills clusters')
plt.legend(loc='upper left')
plt.show()
```

```python
table2 = pd.pivot_table(df,values='Teamwork',\
                    index=['label'],
                     aggfunc='count')
table2.rename(columns={'Teamwork':'count'}, inplace=True)
table2
```

```python
table2.plot.bar()
plt.title('cluster size')
plt.ylabel('count')
plt.xticks(rotation=0)
```
